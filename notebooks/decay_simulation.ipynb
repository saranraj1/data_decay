```python
import pandas as pd
import numpy as np
import os

# Load the unseen production data
df_future = pd.read_csv('../data/processed/future_unseen.csv')
features_to_monitor = ['median_income', 'housing_median_age', 'total_rooms']

# Create a copy for each scenario
df_sensor_drift = df_future.copy()
df_concept_drift = df_future.copy()
```

```python
# Introduce 15% Gaussian noise to 'housing_median_age'
noise_factor = 0.15
noise = np.random.normal(0, df_future['housing_median_age'].std(), len(df_future))
df_sensor_drift['housing_median_age'] += noise * noise_factor

print("‚ö†Ô∏è Scenario A: Sensor Noise injected into 'housing_median_age'.")
```

```python
# Simulate 30% income inflation
df_concept_drift['median_income'] = df_future['median_income'] * 1.30

print("‚ö†Ô∏è Scenario B: Concept Drift (Inflation) injected into 'median_income'.")
```

```python
processed_path = '../data/processed/'
df_sensor_drift.to_csv(f'{processed_path}future_sensor_decay.csv', index=False)
df_concept_drift.to_csv(f'{processed_path}future_concept_drift.csv', index=False)

print("üíæ Decayed datasets saved to /data/processed/")
```
